Import the "Avengers" data from the [FiveThirtyEight data repository on GitHub](https://github.com/fivethirtyeight/data/blob/master/avengers/avengers.csv).

```r
library('readr')
library('tidyverse')
```

```r
df = read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/avengers/avengers.csv')
```

```r
df %>% head
```

Make an ECDF plot of `Appearances`.

```r
p = ggplot(data=df, aes(x=Appearances))
p = p + stat_ecdf(geom='point')
p
```
