Using the following data,

```r
library('tidyverse')

# From http://stackoverflow.com/questions/1181060
stocks = tibble(
  time = as.Date('2009-01-01') + 0:999,
  X = rnorm(1000, 0.35, 1),
  Y = rnorm(1000, 0, 2),
  Z = rnorm(1000, 2, 4)
)

stocks %>% print
```

Re-create the following figure:

(*Hints: use* `gather` *to re-shape the data.*)

<img src="figs/stock_gather.png" style="width: 500px;" />

```r
stocks.wide = stocks %>% gather(stock, price, -time)
stocks.wide %>% print

p = ggplot(stocks.wide, aes(x=stock, y=price))
p = p + geom_point(alpha=0.25, position=position_jitter(h=0, w=0.25))
p
```
