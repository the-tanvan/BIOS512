1. Labelling points

```r
# ggrepel used to repel text labels away from one another.
library('tidyverse')
library('ggrepel')

filtered.data = mpg %>% filter(displ > 5, hwy > 21)  # DRY
```

## Labelling Points

```r
options(repr.plot.res = 120, repr.plot.height = 3, repr.plot.width = 4)
p = ggplot(mpg, aes(x = displ, y = hwy))
p = p + geom_point()
p = p + geom_text(data=mpg %>% filter(displ > 5, hwy > 21), 
              aes(label = manufacturer))
p
```

```r

p + geom_text_repel(
        data = filtered.data,
        nudge_y = 36 - filtered.data$hwy,
        segment.size = 0.2,
        size = 3,
        segment.color = "grey50",
        aes(label = manufacturer)
    )
p
```

## Legends, axes, titles, themes

```r
options(repr.plot.res = 120, repr.plot.height = 3, repr.plot.width = 6)
p = ggplot(mpg, aes(x = displ, y = hwy, color = class))
p = p + geom_point()
p = p + geom_text_repel(
            data = filtered.data,
            nudge_y = 36 - filtered.data$hwy,
            segment.size = 0.2,
            size = 3,
            segment.color = "grey50",
            aes(label = manufacturer),
        )
p = p + theme_light() 
p = p + labs(y='Highway Mileage', x='Displacement', color='class')
p = p + theme(legend.text=element_text(face='bold'),
             axis.title.y=element_text(color='blue'),
             axis.text.x=element_text(color='red'),
             legend.position='bottom')
p = p + ggtitle('This is our title')
p
```
