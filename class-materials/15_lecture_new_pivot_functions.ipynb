# `pivot_longer` versus `gather`

```r
?pivot_longer
# pivot_longer is brand-new tidyr function that replaces gather
# pivot_longer(data, cols, names_to="name"...)
# data =
# cols = 
# names_to="name" =
```

```r
billboard %>% print
```

```r
billboard %>%
    gather(key, value, -artist, -track, -date.entered)
```

```r
# How to print out column names
billboard %>% names
# can also use
# billboard %>% colnames

# How to print out number of columns
billboard %>% names %>% length
```

```r
# Creates key column
# Creates value column
# Remove cols using one_of function (from dplyr package) which selects one_of the columns of the provided list
# The minus indicates to remove

billboard %>%
    pivot_longer(names_to = 'key', values_to = 'value', 
                 cols = -one_of('track', 'artist', 'date.entered'))

# The following syntax does not work with this function (unlike gather function)
# billboard %>%
#     pivot_longer(names_to = 'key', values_to = 'value', -artist, -track, -date.entered)
```

# `pivot_wider` versus `spread`

```r
iris.long = iris %>%
    rownames_to_column() %>%
    pivot_longer(-one_of('Species', 'rowname'), 
                 names_to = 'key', 
                 values_to = 'value')

iris.long %>% print
```

```r
iris.long %>% spread(key, value) %>% head(5)
```

```r
iris.long %>% pivot_wider(names_from = 'key', id_cols = c('rowname', Species)) %>% head(5)
```

```r
# Alternatively
iris.long %>% pivot_wider(names_from = 'key') %>% head (5)
```

# Select functions

- `one_of`
- `matches`
- `contains`
- `starts_with` and `ends_with`
- `everything`

```r
billboard %>%
    pivot_longer(names_to = 'week', values_to = 'rank', cols = starts_with('wk')) %>% head(5)
```

```r
billboard %>%
    pivot_longer(names_to = 'week', values_to = 'rank', cols = matches('wk')) %>% head(5)
```
